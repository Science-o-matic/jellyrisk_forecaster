```python
import pandas as pd
from datetime import datetime, timedelta
from glob import glob
```

# OCEANCOLOUR_MED_CHL

```python
files = sorted(glob('MyOcean/Historical/Preprocessed/dataset-oc-med-chl-multi_cci-l3-chl_4km_daily-rep-v02_*-CHL.csv')) + \
        ['MyOcean/Historical/Preprocessed/dataset-oc-med-chl-modis_a-l3-chl12_1km_daily-rt-v02_2013-06-03_2013-10-23-CHL.csv']
files
```

```python
chl = pd.read_csv('MyOcean/Historical/Preprocessed/dataset-oc-med-chl-multi_cci-l3-chl_4km_daily-rep-v02_2007-05-23_2007-09-18-CHL.csv')
```

```python
chl
```

```python
pd.isnull(chl.CHL).sum()
```

```python
def fix_point_date(df):
    epoch = datetime(1981, 1, 1)
    df['Point.date'] = df['Point.date'].apply(lambda x: (epoch + timedelta(seconds=x)).strftime('%Y%m%d'))
    return df
```

```python
dfs = map(pd.read_csv, files)
dfs = map(fix_point_date, dfs)
chl = pd.concat(dfs, ignore_index=True)
```

```python
chl.to_csv('MyOcean/Historical/Preprocessed/OCEANCOLOUR_MED_CHL-2007-2013.csv', index=False)
```

```python
pd.isnull(chl.CHL).sum()
```

# MEDSEA_REANALYSIS_BIO_006_007

```python
vars = ['chl', 'nit', 'pcb', 'pho', 'dox', 'npp']
files = map(lambda var: sorted(glob('MyOcean/Historical/Preprocessed/myov04-med-ogs-bio-rean_*-{}.csv'.format(var))), vars)
files
```

```python
dfs = [map(pd.read_csv, files_) for files_ in files]
```

```python
dfs = [pd.concat(dfs_vars) for dfs_vars in dfs]
```

```python
df = pd.concat([dfs[0], dfs[1]['nit'], dfs[2]['pcb'], dfs[3]['pho'], dfs[4]['dox'], dfs[5]['npp']], axis=1)
```

```python
epoch = datetime(1970, 1, 1)
df['Point.date'] = df['Point.date'].apply(lambda x: (epoch + timedelta(seconds=x)).strftime('%Y%m%d'))
```

```python
df.to_csv('MyOcean/Historical/Preprocessed/MEDSEA_ANALYSIS_FORECAST_BIO_006_007-2007-05-01_2010-10-01.csv', index=False)
```
