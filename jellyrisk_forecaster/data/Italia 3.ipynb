```python
import pandas as pd
import numpy as np
```

# Get unique beaches

```python
obs = pd.read_excel('Italia/datab j 09-14 completo per descriptive_graphs.xlsx')
obs.info()
```

```python
beaches = pd.DataFrame()
beaches['lon'] = obs['x.long'].str.replace(',', '.').astype(np.float)
beaches['lat'] = obs['y.lat'].str.replace(',', '.').astype(np.float)
beaches
```

```python
beaches_uniq= beaches.drop_duplicates()
beaches_uniq.info()
```

```python
beaches_uniq = beaches_uniq.reset_index(drop=True)
```

```python
beaches_uniq.to_csv('Italia/Italy_beaches_3.csv', index_label='beach')
```

# Add beach id to obs file

```python
obs = pd.read_excel('Italia/datab j 09-14 completo per descriptive_graphs.xlsx')
```

```python
obs['x.long'] = beaches['lon']
obs['y.lat'] = beaches['lat']
obs = obs.merge(beaches_uniq.reset_index().rename(columns={'lon': 'x.long', 'lat': 'y.lat'}),
                on=['x.long', 'y.lat']).rename(columns={'index': 'Site_ID'})
obs
```

```python
salinity = pd.read_csv('MyOcean/Italy/Total/italy-2009-2014-myov04-med-ingv-sal-an-fc-3.csv',
                       parse_dates=['Point.date'], infer_datetime_format=True)
```
